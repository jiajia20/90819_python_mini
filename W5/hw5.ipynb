```python
import requests
import json
import pandas as pd
```

```python
list = ['Python', 'Data Science', 'Data Analysis', 'Machine Learning', 'Deep Learning']

bigTable = pd.DataFrame()
for item in list:
    url = 'https://www.googleapis.com/books/v1/volumes?q=' + item
    response = requests.get(url)
    if response.status_code == 200:
        data = json.loads(response.text)
        temp1 = pd.json_normalize (data['items'])
        #Extract just the 'volumeInfo.title' and 'volumeInfo.authors' columns from temp1 into a DataFrame named temp2
        temp2 = temp1[['volumeInfo.title','volumeInfo.authors']]
        #then, use copy( ) to copy temp2 into a DataFrame named ‘table’. 
        table = temp2.copy()
        #rename table’s columns as 'Title' and 'Authors'.
        table.rename(columns={'volumeInfo.title': 'Title', 'volumeInfo.authors': 'Authors'}, inplace= True)
        #append table to bigTable
        bigTable = bigTable.append(table, ignore_index=True)
```
