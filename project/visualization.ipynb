```python
# create a visualzation of the csv file

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# read the csv file
df_news = pd.read_csv('news_volume.csv')
# chage volume column title to news
df_news = df_news.rename(columns={'volume': 'news'})

# Twitter file
df_twitter = pd.read_csv('length_list.csv')
# chage volume column title to twitter
df_twitter = df_twitter.rename(columns={'length': 'twitter'})

# sort the data by date
df_news = df_news.sort_values(by=['date'])
df_twitter = df_twitter.sort_values(by=['date'])
```

```python
type(df_news['date'][1])
```

```python
# join the two dataframes by date with full outer join
df = pd.merge(df_news, df_twitter, on='date', how='outer')
# fill the missing values with 0
df = df.fillna(0)
```

```python
df.head()
```

```python
# change the date to datetime format
df['date'] = pd.to_datetime(df['date'])
```
